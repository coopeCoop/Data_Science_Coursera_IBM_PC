### TASK  3: Visualize the relationship between success rate of each orbit type


Next, we want to visually check if there are any relationship between success rate and orbit type.


Let's create a `bar chart` for the sucess rate of each orbit


```python
df.head(10)
```

```python
import pandas as pd

# Assuming you have a DataFrame named 'df' with 'Orbit' and 'Class' columns
mean_class_by_orbit = df.groupby('Orbit')['Class'].mean().reset_index()

# Display the result
mean_class_by_orbit
```

```python
from matplotlib import pyplot as plt

# Assuming mean_class_by_orbit is a DataFrame that includes 'Orbit' and 'Class'
# Calculate the success rate for each orbit
success_rate = mean_class_by_orbit.groupby('Orbit')['Class'].mean()

# Set the figure size for better visibility
plt.figure(figsize=(10, 6))

# Create the bar chart
bars = plt.bar(success_rate.index, success_rate.values, color='steelblue', edgecolor='black')

# Add title and labels
plt.title('Success Rate by Orbit')
plt.xlabel('Orbit')
plt.ylabel('Success Rate')

# Add values on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom')

# Add a legend if necessary (you may customize this based on your needs)
plt.legend(['Success Rate'])

# Show the plot
plt.show()
```

```python
from matplotlib import pyplot as plt

# Create a line plot for the 'Class' column from the mean_class_by_orbit DataFrame
mean_class_by_orbit['Class'].plot(
    kind='line',          # Specify the type of plot as line
    figsize=(8, 4),      # Set the figure size (width, height)
    title='Class'        # Set the title of the plot
)

# Hide the top and right spines for a cleaner plot appearance
plt.gca().spines[['top', 'right']].set_visible(False)

# Optionally, show the plot if running in an interactive environment
plt.show()
```

Analyze the plotted bar chart to identify which orbits have the highest success rates.

### TASK  6: Visualize the launch success yearly trend


You can plot a line chart with x axis to be <code>Year</code> and y axis to be average success rate, to get the average launch success trend.


The function will help you get the year from the date:


```python
# A function to extract years from the date column
def extract_year(df):
    df['Year'] = df['Date'].apply(lambda x: int(x.split('-')[0]))
    return df

# Apply the function to extract the year and add it to the DataFrame
df = extract_year(df)

# Display the DataFrame to ensure 'Year' column is added
df.head()
```

```python
import matplotlib.pyplot as plt

# Compute the average success rate by year
average_by_year = df.groupby('Year')['Class'].mean()

# Reset the index to have 'Year' as a regular column
average_by_year = average_by_year.reset_index()

# Plot the line chart with Year on the x-axis and Success Rate on the y-axis
plt.figure(figsize=(8, 6))
plt.plot(average_by_year['Year'], average_by_year['Class'], marker='o', linestyle='-')

# Add labels and title
plt.xlabel('Year', fontsize=14)
plt.ylabel('Average Success Rate', fontsize=12)
plt.title('Average Launch Success Rate Yearly Trend', fontsize=14)
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Display the plot
plt.show()
```

you can observe that the sucess rate since 2013 kept increasing till 2020


## Features Engineering


By now, you should obtain some preliminary insights about how each important variable would affect the success rate, we will select the features that will be used in success prediction in the future module.


```python
features = df[['FlightNumber', 'PayloadMass', 'Orbit', 'LaunchSite', 'Flights', 'GridFins', 'Reused', 'Legs', 'LandingPad', 'Block', 'ReusedCount', 'Serial']]
features.head()
```

### TASK  7: Create dummy variables to categorical columns


Use the function <code>get_dummies</code> and <code>features</code> dataframe to apply OneHotEncoder to the column <code>Orbits</code>, <code>LaunchSite</code>, <code>LandingPad</code>, and <code>Serial</code>. Assign the value to the variable <code>features_one_hot</code>, display the results using the method head. Your result dataframe must include all features including the encoded ones.


```python
# Use pd.get_dummies() to perform one-hot encoding on categorical features
features_one_hot = pd.get_dummies(features, columns=['Orbit', 'LaunchSite', 'LandingPad', 'Serial'])

# Display the resulting DataFrame
features_one_hot.head()
```

### TASK  8: Cast all numeric columns to `float64`


Now that our <code>features_one_hot</code> dataframe only contains numbers, cast the entire dataframe to variable type <code>float64</code>


```python
# Use the astype() method to cast all columns to float64
features_one_hot = features_one_hot.astype('float64')

# Display the DataFrame to ensure the conversion
features_one_hot.head()
```
